```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# need to pip install xlrd to read csv files
df = pd.read_excel("datasets/pd4.xlsx", sheet_name='Arkusz1')

df.head()
```

```python
parameters_list = ["p"+str(i+1) for i in range(11)]
parameters_list
```

```python
# Plots
sheets_list = []

for i in range(11):
    for j in range(11):
        if j > i:
            df_tmp = df[ df[parameters_list[i]] != -1 ]
            df_tmp = df_tmp[ df_tmp[parameters_list[j]] != -1 ]
            df_tmp = df_tmp.sort_values(parameters_list[i])
            index = parameters_list[i] + parameters_list[j]
            sheets_list.append([index, df_tmp])
            df_tmp2 = df_tmp[df_tmp['wynik'] == 1] 
            x = df_tmp2[parameters_list[i]].to_numpy()
            y = df_tmp2[parameters_list[j]].to_numpy()
            plt.plot(x,y, 'r', label='Chorzy') # Wykres chorzy
            df_tmp2 = df_tmp[df_tmp['wynik'] == 0] 
            x = df_tmp2[parameters_list[i]].to_numpy()
            y = df_tmp2[parameters_list[j]].to_numpy()
            plt.plot(x,y, 'b', label='Zdrowi') # Wykres chorzy
            plt.title("{}({})".format(parameters_list[i], parameters_list[j]))
            plt.legend()
            plt.xlabel(parameters_list[i])
            plt.ylabel(parameters_list[j])
            plt.savefig('plots/{}_od_{}.png'.format(parameters_list[i], parameters_list[j]))
            plt.close()

            
print(sheets_list[0][0], "\n", sheets_list[0][1])
print(sheets_list[1][0], "\n", sheets_list[1][1])
```

```python
import statistics as stats
from scipy import stats
# Wskazniki statystyczne
f = open("Wyniki.txt", "a")
f.truncate(0)
#f.write("\n\n")
for parameter in parameters_list:
    line = parameter + "\n"
    f.write(line)
    # Chorzy
    line = "Chorzy\n"
    f.write(line)
    df_tmp = df[ df['wynik'] == 1] 
    df_tmp = df_tmp[ df_tmp[parameter] != -1]
    data = df_tmp[parameter].to_numpy()
    wariancja = np.var(data)
    wartosc_oczekiwana = np.mean(data)
    line = "\tWartosc oczekiwna = " + str(wartosc_oczekiwana) + "\n"
    f.write(line)
    line = "\tWariancja = " + str(wariancja) + "\n"
    f.write(line)
    # Zdrowi
    line = "Zdrowi\n"
    f.write(line)
    df_tmp = df[ df['wynik'] == 0] 
    df_tmp = df_tmp[ df_tmp[parameter] != -1]
    data = df_tmp[parameter].to_numpy()
    wariancja = np.var(data)
    wartosc_oczekiwana = np.mean(data)
    line = "\tWartosc oczekiwna = " + str(wartosc_oczekiwana) + "\n"
    f.write(line)
    line = "\tWariancja = " + str(wariancja) + "\n"
    f.write(line)
    
f.write("\n\n Wspolczynniki Pearsona:\n")    
for i in range(len(parameters_list)):
    for j in range(len(parameters_list)):
        if j > i:
            
            f.write("\n" + parameters_list[i] + ", " + parameters_list[j] + "\n")
            df_tmp = df[ df[parameters_list[i]] != -1 ]
            df_tmp = df_tmp[ df_tmp[parameters_list[j]] != -1 ]
            df_tmp = df_tmp.sort_values(parameters_list[i])
            # chorzy
            f.write("Chorzy: \n")
            df_tmp2 = df_tmp[df_tmp['wynik'] == 1] 
            x = df_tmp2[parameters_list[i]].to_numpy()
            y = df_tmp2[parameters_list[j]].to_numpy()
            wsp_pearsona = stats.pearsonr(x, y)
            f.write(str(wsp_pearsona))
            # zdrowi
            f.write("\nZdrowi: \n")
            df_tmp2 = df_tmp[df_tmp['wynik'] == 0] 
            x = df_tmp2[parameters_list[i]].to_numpy()
            y = df_tmp2[parameters_list[j]].to_numpy()
            wsp_pearsona = stats.pearsonr(x, y)
            f.write(str(wsp_pearsona))


f.close()

```
