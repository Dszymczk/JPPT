```python
import numpy as np
import pandas as pd

df = pd.read_excel("datasets/pd4.xlsx", sheet_name='Arkusz1')

df.head()
```

```python
parameters_list = ["p"+str(i+1) for i in range(11)]
parameters_list
```

```python
list_of_dfs = []
for i in range(len(parameters_list)):
    for j in range(len(parameters_list)):
        if j > i:
            new_df = df[df[parameters_list[j]] != -1]
            new_df = df[df[parameters_list[i]] != -1]
            new_df = new_df[[parameters_list[i], parameters_list[j],  "wynik"]].copy()

            list_of_dfs.append(new_df)
    
    
```

```python
list_of_dfs[1]
```

```python
for list in list_of_dfs:
    print(len(list))
    print(list.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score

for list in list_of_dfs:
    #print(list.columns)
    #print(list.columns[0])
    x = list[[list.columns[0], list.columns[1]]]
    y = list['wynik']
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .8)

    my_classifier = tree.DecisionTreeClassifier()
    my_classifier.fit(x_train, y_train)

    predictions = my_classifier.predict(x_test)
    score = accuracy_score(y_test, predictions)
    print("Accuracy for {}{}: ".format(list.columns[0], list.columns[1]), score)
```
